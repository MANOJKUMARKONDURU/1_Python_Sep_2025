```python
#Arrays
# Python does not have in built arrays but python lists can be used instead

# Array is a special variable, which can hold more than one value at a time

#create an array containing car names:
cars = ["Ford", "Mustang", "Pagani", "Benz"]

#An Array can hold many values under a single name, and can access the values by reffering to the index number
x = cars[0]
print(x)

# add a value to array:
cars.append("Tyota")
print(cars)

#modify a value to array
cars[3] = "Bugati"
print(cars[3])


# Lenght of an Arrray - len()
# len() method to return the length of an array
x = len(cars)
print(x)   #4

#the lenght of an array is always one more than the highest array index


print()
# Looping through arrays -   for  in   to loop through all the elements of an array

for x in cars:
  print(x)

print()


#Adding Array elements - append() method to add an element to an array
cars.append("Lambo")
print(cars)

#Removing Array element - pop() to remeove an element from the array with index number
cars.pop(3)
print(cars)

#remove() to remove an element in an array with the specific value
cars.remove("Tyota")
print(cars)

```

```python
# 3. Reverse an array without using built-in functions ([::-1] or reverse()).


# 1st approachuses extra memory - for creating an extra array
num = [1, 2, 3, 4, 5]

rev = []
for i in range(len(num) -1, -1, -1):   #range(start, stop, step)
  rev.append(num[i])
print(f"Rverse an array: {rev}")



# 2nd approach
num = [1, 2, 3, 4, 5]
start = 0
end = len(num) - 1
while start < end:
  num[start], num[end] = num[end], num[start]
  start += 1
  end -= 1
print(f"Reverse an array: {num}")
```

```python
#6.Rotate an array by k positions to the right (e.g., [1,2,3,4,5] rotated by 2 â†’ [4,5,1,2,3]).

def rotated_array(arr, k):
  n = len(arr)
  k = k % n
  rotated = arr[-k: ] + arr[:-k] # last k + first n - k
           #arr[-2:] = [5,4],  arr[:-k] = [1,2,3]
  return rotated

arr = [1, 2, 3, 4, 5]
k = 2
print(rotated_array(arr, k))
```

```python
# 8.Merge two sorted arrays into a single sorted array.
first_array = [1, 2, 3, 4, 5,10]
second_array = [4, 7, 8, 9]

merged_array = first_array + second_array
merged_array.sort()
print(merged_array)


new_merged_array = []

i = 0   #pointer for first_array
j = 0   #pointer for second_array

#Step 1: Merge while both have elements
while i < len(first_array) and j < len(second_array):
  if first_array[i] < second_array[j]:     #if  (i)th element of first_array is leass than (j)th element of the second_array
    new_merged_array.append(first_array[i])  # append the (i)th element to the new_merged_array
    i += 1                                  # move forward
  else:
    new_merged_array.append(second_array[j])     # else appened (j)th element to new_merged_array
    j +=1                                    #move forward

#Step 2: appened the remaining(left over) elements
while i < len(first_array):
  new_merged_array.append(first_array[i])
  i += 1

#step 3: appened the remaining(left over) elements
while j < len(second_array):
  new_merged_array.append(second_array[j])
  j += 1

#Print the new_merged_array
print(f"The new Merged array of two sorted array: {new_merged_array}")
```
